1. Анализ данных

```python
import pandas as pd
#import nona.nona
import numpy as np
data = pd.read_csv('assignment_1.csv')
data.head()
```

```python
# Разбиваем столбец Cabin 
data['Deck'] = data['Cabin'].str.split('/').str.get(0)
data['Side'] = data['Cabin'].str.split('/').str.get(2)
data = data.drop(columns=['Cabin'])
data.head()
```

```python
data.isnull().sum() #пропущенные значения 
```

```python
#from nona.nona import nona
# Заполняем пропущенные значения
#nona(data=data, algreg=make_pipeline(StandardScaler(with_mean=False), Ridge(alpha=0.1)), algclass=RandomForestClassifier(max_depth=2, random_state=0))
data.fillna(method='ffill', inplace=True)  #простой и удобный способ заполнить пропуски значениями предшествующих объектов
data.head()
```

```python
data.isna().sum() #ПРОВЕРКА
```

```python
# организация признаков 
data = data.drop(columns=['PassengerId', 'Name', 'HomePlanet','Destination', 'Age', 'VIP','RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck','Name'])
data.head()
```

```python
# from sklearn import preprocessing
# label_encoder = preprocessing.LabelEncoder()
# data['Deck'] = label_encoder.fit_transform(data['Deck'])
# data['Side'] = label_encoder.fit_transform(data['Side'])

# Кодируем категориальные признаки, в числовой формат
data = pd.get_dummies(data, columns=['Deck', 'Side'])
data.head()
```

2. Обучение модели

```python
# X=data[['Cabin', 'CryoSleep']]
X=data.drop(columns=['Transported'])
y=data['Transported']
from sklearn.model_selection import train_test_split
#размеры выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train) 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred = model.predict(X_test)                     # Предсказание значений
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1-score:', f1_score(y_test, y_pred))       # Оценка качества модели
```
